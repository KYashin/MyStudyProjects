# Телеком — задача проекта

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

**Описание услуг:**

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

**Описание данных:**

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

### Создание целового признака:

Напишем функцию:


```python
def create_target(end_date):   # функция для создания столбца, содержащего целевой признак. Если строка в столбце 'end_date'
    if 'No' in end_date:       # содержит строку 'No', то столбцу-целевому признаку присваивается 0, иначе 1.
        return 0
    else:
        return 1
    
```

Применим функцию к столбцу `end_date`:

### `Pipeline`

#### Распределение столбцов по типам

Распределим столбцы по трём группам: категориальные небинарные признаки, категориальные бинарные признаки, и числовые признаки.

```python
cat_columns = ['type', 'payment_method']  # столбцы, содержащие небинарные категориальные признаки

binary_columns = features.drop(
    [
        'senior_citizen',
        'type',
        'payment_method',
        'monthly_charges',                # столбцы, содержащие бинарные категориальные признаки
        'total_charges',
        'diff_days'
    ], axis=1).columns

num_columns = ['total_charges', 'monthly_charges']       # столбцы, содержащие числовые признаки 'monthly_charges',
```

```python
features_train.columns.to_list()
```

### Создание `Pipline`

Создадим объект типа `ColumnTransformer` и добавим в него те шаги, которые будем делать c определенным типом данных в столбцах, которые я выделил заранее:

```python
columns_pipline = ColumnTransformer(
    [('num_cols', StandardScaler(), num_columns),
     ('category_encoder',
      OrdinalEncoder(handle_unknown='use_encoded_value',
                     unknown_value=-1), cat_columns),
     ('binary_encoder', OneHotEncoder(
         drop='first', handle_unknown='ignore'), binary_columns)],
    remainder='passthrough')
columns_pipline
```

С помощью функции `make_pipline`, преобразуем `ColumnTransformer` в Pipeline:

```python
pipeline = make_pipeline(columns_pipline)
```

Добавим модель в наш пайплайн, в данном случае Логистическую регрессию:

```python
pipeline.steps.append(('clf', LogisticRegression(class_weight='balanced', tol=0.01, random_state=270323)))
pipeline
```

Выделим отдельно категориальные признаки для далнейшего указания их в параметрах `CatBoostClassifier`:

```python
cat_features = features_train.select_dtypes('object').columns.to_list()
```

```python
cat_features
```

Распишем модели и параметры для перебора, чтобы выявить лучшую:

```python
params = [
    {
        'clf__solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
        #    'clf__penalty': ['l2', 'l1'],
        'clf__C': [0.1, 1, 10, 100],
        'clf__max_iter': range(1, 1001, 100),
        'columntransformer__binary_encoder': [OneHotEncoder()],
        'columntransformer__category_encoder': [OneHotEncoder()]
    },
    {
        'clf': [RandomForestClassifier(class_weight='balanced', random_state=270323)],
        'clf__max_depth': range(1, 20),
        'clf__n_estimators': range(10, 501, 10),
        'clf__min_samples_leaf': range(1, 8),
        'clf__min_samples_split': range(2, 10, 2),
        'clf__criterion': ['entropy', 'gini'],
        'columntransformer__binary_encoder': [OneHotEncoder(), OrdinalEncoder()],
        'columntransformer__category_encoder': [OneHotEncoder(), OrdinalEncoder()],
        'columntransformer__num_cols': ['passthrough']
    },
    {
        'clf': [CatBoostClassifier(cat_features=cat_features, class_weights='balanced', verbose=0, random_state=270323)],
        'clf__iterations': range(10, 101, 10),
        'clf__n_estimators': range(10, 601, 10),
        'clf__max_depth': range(1, 20),
        'clf__depth': [4, 5, 6, 7, 8, 9, 10],
        'clf__learning_rate' : [0.01, 0.02, 0.03, 0.04],
        'columntransformer__binary_encoder': [None],
        'columntransformer__category_encoder': [None],
        'columntransformer__num_cols': ['passthrough']
    }
]
```

Инициализируем `RandomizedSearchCV` в отдельную переменную, укажем все необходимые параметры:

```python
random_src = RandomizedSearchCV(pipeline,
                                params,
                                cv=3,
                                n_iter=500,
                                n_jobs=-1,
                                random_state=270323,
                                scoring='roc_auc',
                                verbose=False)
```

Обучим `RandomizedSearchCV` на обучающей выборке: 

```python
random_src.fit(features_train, target_train)
```

Выведем лучший `Pipline`:

```python
random_src.best_estimator_
```

Выведем гиперпараметры лучшей модели:

```python
random_src.best_params_
```

```python
print(f'ROC_AUC лучшей модели на кросс-валидации: {round(random_src.best_score_, 3)}')
```

### Создание `Pipline`

```python
columns_pipline = ColumnTransformer(
    [('num_cols', StandardScaler(), num_columns),
     ('category_encoder',
      OrdinalEncoder(handle_unknown='use_encoded_value',
                     unknown_value=-1), cat_columns),
     ('binary_encoder', OneHotEncoder(
         drop='first', handle_unknown='ignore'), binary_columns)],
    remainder='passthrough')
columns_pipline
```

```python
pipeline = make_pipeline(columns_pipline)
```

```python
pipeline.steps.append(('clf', LogisticRegression(class_weight='balanced', tol=0.01, random_state=270323)))
pipeline
```
