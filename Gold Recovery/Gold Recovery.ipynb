### Проверка эффективности обогащения в обучающей выборке для `rougher.output.recovery`

**Посчитаем MAE вручную.**

Для этого добавим в обучающую выборку столбец с расчётной эффективностью обогащения:

```python
gold_recovery_train_new['rougher.calculation.recovery'] = (
    gold_recovery_train_new['rougher.output.concentrate_au'] *
    (gold_recovery_train_new['rougher.input.feed_au'] -
     gold_recovery_train_new['rougher.output.tail_au'])) / (
         gold_recovery_train_new['rougher.input.feed_au'] *
         (gold_recovery_train_new['rougher.output.concentrate_au'] -
          gold_recovery_train_new['rougher.output.tail_au'])) * 100
```

<font color='blue'><b>Комментарий ревьюера: </b></font> ✔️\
<font color='green'>Расчёт верный!</font>

```python
MAE_manually = np.abs((gold_recovery_train_new['rougher.output.recovery'] -
                       gold_recovery_train_new['rougher.calculation.recovery'])).mean()
MAE_manually
```

MAE, посчитанная вручную, получилась близка к 0.

Посчитаем MAE с помощью функции из `sklearn` для сравнения (и ради интереса):

```python
MAE = mean_absolute_error(gold_recovery_train_new['rougher.calculation.recovery'],
                          gold_recovery_train_new['rougher.output.recovery'])
MAE
```

Получилось, то же самое значение. MAE рассчитано верно.

**Вывод:** `MAE≈0` это значит, что эффективность обогащения рассчитана правильно.

#### Работа с выбросами

Напишем функцию:

```python
def sigma_outlier(feature, y=False, threshold=3):
    sd = np.std(feature, axis=0)
    mean = np.mean(feature, axis=0)
    upper = mean + threshold * sd
    lower = mean - threshold * sd
    idx = np.where((feature < lower) | (feature > upper))[0]
    mask = np.ones(feature.shape[0], dtype=int)
    mask[idx] = -1
    if y:
        return feature[mask == 1], y[mask == 1]
    return feature[mask == 1]
```

Применим функцию к столбцам с суммами:

```python
gold_recovery_train_new['sum_rougher_input'] = sigma_outlier(
    gold_recovery_train_new['sum_rougher_input'])
gold_recovery_train_new['sum_rougher_output'] = sigma_outlier(
    gold_recovery_train_new['sum_rougher_output'])
gold_recovery_train_new['sum_final_output'] = sigma_outlier(
    gold_recovery_train_new['sum_final_output'])
```

## Модель

### Функция для вычисления итоговой sMAPE

Функция для вычисления sMAPE на разных стадиях:

```python
def intermediate_smape(target, prediction):
    intermediate_smape = (sum(
        ((abs(target - prediction)) /
         (abs(target) + (abs(prediction)) * .5)) * 100)) / len(target)
    return intermediate_smape
```

Напишем функцию для вычисления итоговой sMAPE:

```python
def smape(rougher_smape, final_smape):
    aggregate_smape = .25 * rougher_smape + .75 * final_smape
    return aggregate_smape
```
